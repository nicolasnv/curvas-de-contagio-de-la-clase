```python
import numpy as np
```

```python
k=np.log(19999/((20000/345)-1))/50
```

```python
def c(t):
  return 20000/(1+19999*np.exp(-0.117*t))
```

```python
t = np.arange(0,180,0.01)
```

```python
c(t)
```

```python
dcdt=k*c(t)*(1-(c(t)/20000))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plt.plot(t, c(t),c="red",linestyle=":",label="Contagios")
plt.axvline(np.log(19999)/k,)

plt.title("Curva de contagios",fontsize= 20)
plt.xlabel("Tiempo",c="blue",fontsize= 20)
plt.ylabel("Contagios",c="green",fontsize= 20)

plt.legend(loc="upper left")

 
```
